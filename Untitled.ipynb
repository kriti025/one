```python
##On some of these columns, a value of zero does not make sense and thus indicates missing value.
```

```python
col=['Glucose' ,'BloodPressure' ,'SkinThickness', 'Insulin' ,'BMI']
```

```python
for i in col:
  data[i].replace(0,data[i].mean(),inplace=True)
```

```python
X= data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
X
```

```python
y=data['Outcome']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
plt.scatter(y_test,predictions)
```

```python
p=logmodel.predict([[5,116,74,0,0,25.6,0.201,30]])
```

```python
p[0]
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

```python
predictions = svc_model.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Train Set Accuracy:"+str(accuracy_score(y_train,svc_model.predict(X_train))*100))
print("Test Set Accuracy:"+str(accuracy_score(y_test,svc_model.predict(X_test))*100))
```
